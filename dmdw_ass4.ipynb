```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="https://raw.githubusercontent.com/18cse169/aman18cse169/main/indian_food.csv"
data=pd.read_csv(path)
data.head()
```

```python
data.dropna(axis=0,inplace=True)
```

```python
data.head()
```

```python
sns.set(style="darkgrid")
sns.regplot(x=data['cook_time'],y=data['prep_time'])
```
